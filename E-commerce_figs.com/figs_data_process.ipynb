```python
import pandas as pd
df = pd.read_excel("scraped_data.xlsx") 
df.head()
```

```python
df = df.drop(columns=['Product URL'])
df = df.drop(columns=['Product Category'])
df = df.dropna(subset=['Product Name'])
df['Original Price'] = df['Original Price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['Discounted Price'] = df['Original Price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['Discounted Limited Edition Colors Price'] = df['Original Price'] * 0.65
df.head()
```

```python
df.head()
```

```python
df.to_csv("figs_file.csv", index=False)
```
